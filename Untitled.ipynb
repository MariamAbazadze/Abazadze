```python
import sqlite3
conn = sqlite3.connect("netflix_mai.sqlite")
cursor = conn.cursor()
print("·Éô·Éê·Éï·É®·Éò·É†·Éò ·É¨·Éê·É†·Éõ·Éê·É¢·Éî·Éë·Éò·Éó ·Éì·Éê·Éõ·Éß·Éê·É†·Éì·Éê")
```

```python
cursor.execute("SELECT title, release_year FROM netflix WHERE release_year = 2020")
rows = cursor.fetchall()
cursor.execute("PRAGMA table_info(netflix)")
columns = cursor.fetchall()
print("·É™·ÉÆ·É†·Éò·Éö·Éò·É° ·É°·Éï·Éî·É¢·Éî·Éë·Éò:")
for col in columns:
    print(col)
print(" 2020 ·É¨·Éö·Éò·É° ·É§·Éò·Éö·Éõ·Éî·Éë·Éò:")
for row in rows:
    print(row)
# ·Éô·Éù·Éì·Éò ·Éô·Éò·Éó·ÉÆ·É£·Éö·Éù·Éë·É° 2020 ·É¨·Éî·Éö·É° ·Éí·Éê·Éõ·Éù·É°·É£·Éö ·É§·Éò·Éö·Éõ·Éî·Éë·É° ·Éì·Éê ·Éë·Éî·É≠·Éì·Éê·Éï·É° ·Éõ·Éê·Éó.
```

```python
title = input("·É®·Éî·Éò·Éß·Éï·Éê·Éú·Éî·Éó ·É§·Éò·Éö·Éõ·Éò·É° ·É°·Éê·ÉÆ·Éî·Éö·Éò: ")
year = int(input("·É®·Éî·Éò·Éß·Éï·Éê·Éú·Éî·Éó ·Éí·Éê·Éõ·Éù·É®·Éï·Éî·Éë·Éò·É° ·É¨·Éî·Éö·Éò: "))

cursor.execute("INSERT INTO netflix (title, release_year) VALUES (?, ?)", (title, year))
conn.commit()

print("‚úÖ ·É©·Éê·Éú·Éê·É¨·Éî·É†·Éò ·É¨·Éê·É†·Éõ·Éê·É¢·Éî·Éë·Éò·Éó ·Éì·Éê·Éî·Éõ·Éê·É¢·Éê")
# ·Éô·Éù·Éõ·Éî·Éú·É¢·Éê·É†·Éò: ·Éî·É° ·Éô·Éù·Éì·Éò ·Éê·Éõ·Éê·É¢·Éî·Éë·É° ·Éê·ÉÆ·Éê·Éö ·É§·Éò·Éö·Éõ·É° ·É™·ÉÆ·É†·Éò·Éö·É®·Éò ·Éõ·Éù·Éõ·ÉÆ·Éõ·Éê·É†·Éî·Éë·Éö·Éò·É° ·É®·Éî·Éß·Éï·Éê·Éú·Éò·Éö·Éò ·Éõ·Éù·Éú·Éê·É™·Éî·Éõ·Éî·Éë·Éò·Éó.
```

```python
old_title = input("·É®·Éî·Éò·Éß·Éï·Éê·Éú·Éî·Éó ·É´·Éï·Éî·Éö·Éò ·É°·Éê·ÉÆ·Éî·Éö·Éò: ")
new_title = input("·É®·Éî·Éò·Éß·Éï·Éê·Éú·Éî·Éó ·Éê·ÉÆ·Éê·Éö·Éò ·É°·Éê·ÉÆ·Éî·Éö·Éò: ")

cursor.execute("UPDATE netflix SET title = ? WHERE title = ?", (new_title, old_title))
conn.commit()

print("‚úÖ ·É°·Éê·ÉÆ·Éî·Éö·Éò ·Éí·Éê·Éú·Éê·ÉÆ·Éö·Éì·Éê")
# ·Éô·Éù·Éõ·Éî·Éú·É¢·Éê·É†·Éò: ·Éõ·Éù·É™·Éî·Éõ·É£·Éö·Éò ·Éô·Éù·Éì·Éò ·É™·Éï·Éö·Éò·É° ·É§·Éò·Éö·Éõ·Éò·É° ·É°·Éê·ÉÆ·Éî·Éö·É° ·Éë·Éê·Éñ·Éê·É®·Éò.

```

```python
delete_title = input("·É®·Éî·Éò·Éß·Éï·Éê·Éú·Éî·Éó ·É¨·Éê·É°·Éê·É®·Éö·Éî·Éö·Éò ·É§·Éò·Éö·Éõ·Éò·É° ·É°·Éê·ÉÆ·Éî·Éö·Éò: ")

cursor.execute("DELETE FROM netflix WHERE title = ?", (delete_title,))
conn.commit()

print(" ·É©·Éê·Éú·Éê·É¨·Éî·É†·Éò ·É¨·Éê·Éò·É®·Éê·Éö·Éê")
# ·Éô·Éù·Éì·Éò ·É®·Éö·Éò·É° ·Éõ·Éò·Éó·Éò·Éó·Éî·Éë·É£·Éö·Éò ·É°·Éê·ÉÆ·Éî·Éö·Éò·É° ·Éõ·É•·Éù·Éú·Éî ·É§·Éò·Éö·Éõ·É°.
```

```python
import matplotlib.pyplot as plt

cursor.execute("SELECT release_year, COUNT(*) FROM netflix GROUP BY release_year")
data = cursor.fetchall()

years = [row[0] for row in data]
counts = [row[1] for row in data]

plt.bar(years, counts)
plt.title("·É¨·Éö·Éî·Éë·Éò·É° ·Éõ·Éò·ÉÆ·Éî·Éì·Éï·Éò·Éó ·É§·Éò·Éö·Éõ·Éî·Éë·Éò·É° ·É†·Éê·Éù·Éì·Éî·Éú·Éù·Éë·Éê")
plt.xlabel("·É¨·Éî·Éö·Éò")
plt.ylabel("·É†·Éê·Éù·Éì·Éî·Éú·Éù·Éë·Éê")
plt.show()
# ·Éî·É° ·Éë·Éê·É†-·Éì·Éò·Éê·Éí·É†·Éê·Éõ·Éê ·Éê·É©·Éï·Éî·Éú·Éî·Éë·É° ·É†·Éê·Éõ·Éì·Éî·Éú·Éò ·É§·Éò·Éö·Éõ·Éò ·Éí·Éê·Éõ·Éù·Éï·Éò·Éì·Éê ·É°·ÉÆ·Éï·Éê·Éì·Éê·É°·ÉÆ·Éï·Éê ·É¨·Éî·Éö·É°.
```
